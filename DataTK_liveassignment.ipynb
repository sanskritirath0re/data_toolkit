```python
#3.Explain the following termsg
#a.The difference in np.array, np.asarray and np.asanyarray.
#b.The difference between Deep copy and shallow copy.
#answers:(a)
#numpy.array: This function creates an array from an existing iterable, such as a list or tuple.
original_list = [1, 2, 3]
numpy_array = np.array(original_list)
print(numpy_array)
#numpy.asarray: Similar to numpy.array, numpy.asarray converts input to an array. If the input is already an array,it returns that array otherwise it creates a new one.
original_tuple = (4, 5, 6)
numpy_array = np.asarray(original_tuple)
print(numpy_array)
#numpy.asanyarray:Similar to np.asarray, but it passes through subclasses of np.ndarray without copying.
original_list = [7,8,9]
numpy_array = np.asanyarray(original_list)
#answers:(b)
#shallow_copy=Copies the structure of the object, but not the elements.
#deep_copy=Copies both the structure and the elements
original_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
shallow_copied_list = copy.copy(original_list)
shallow_copied_list[0][0] = 'a'
print("Original list:", original_list)
print("Shallow copied list:", shallow_copied_list)
original_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
deep_copied_list = copy.deepcopy(original_list)
deep_copied_list[0][0] = 'a'
print("Original list:", original_list)
print("Deep copied list:", deep_copied_list)
```

```python
#14.Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
data = {
    'Date': pd.date_range(start='2024-01-01', periods=365),
    'Temperature': np.random.randint(10, 30, 365),
    'Humidity': np.random.randint(40, 80, 365)
}
df = pd.DataFrame(data)
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity'.
# b) Label the x-axis as 'Date'.
# c) Set the title of the plot as 'Temperature and Humidity Over Time'.
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
#15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
data = np.random.normal(loc=0, scale=1, size=1000)
# a) Plot a histogram of the data with 30 bins.
plt.hist(data, bins=30, density=True, alpha=0.5, color='b', edgecolor='black', label='Histogram')
# b) Overlay a line plot representing the normal distribution's probability density function (PDF).
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')
# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
# d) Set the title of the plot as 'Histogram with PDF Overlay'.
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

```python
#17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
quadrant = np.zeros_like(x, dtype=int)
quadrant[(x >= 0) & (y >= 0)] = 1  # Quadrant I
quadrant[(x < 0) & (y >= 0)] = 2   # Quadrant II
quadrant[(x < 0) & (y < 0)] = 3    # Quadrant III
quadrant[(x >= 0) & (y < 0)] = 4   # Quadrant IV
data = {'x': x, 'y': y, 'quadrant': quadrant}
sns.scatterplot(data=data, x='x', y='y', hue='quadrant', palette='Set1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.show()
```

```python
#18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'
output_notebook()
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y')
p.line(x, y, legend_label="sin(x)", line_width=2)
p.grid.grid_line_alpha = 0.3
show(p)
```

```python
#19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart.
# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, len(categories))

# Create the figure
p = figure(x_range=categories, height=350, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="hover", tooltips=[("Category", "@x"), ("Value", "@top")])

# Add bars with colors based on values
p.vbar(x=categories, top=values, width=0.9, fill_color='#FF5733', line_color='white')

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Set the title
p.title.align = 'center'

# Remove grid lines
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Show the plot
show(p)
```

```python
#20.Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = np.random.randn(100)
trace = go.Scatter(x=x, y=y, mode='lines', name='Random Data')
layout = go.Layout(
    title='Simple Line Plot',
    xaxis=dict(title='X-axis'),
    yaxis=dict(title='Y-axis')
)
fig = go.Figure(data=[trace], layout=layout)
pio.show(fig)
```
